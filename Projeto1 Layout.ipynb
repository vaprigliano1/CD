## Introdução


O projeto a seguir apresentará uma analise sobre o acesso a internet em dois estados brasileiros para os anos de 2013, 2014 e 2015. Para fins de estudo, foram selecionadas regiões específicas e domicilios com rendas per capitas especéficas. A população estudada se econtra nos estados de São Paulo e Bahia, com renda per capita de até 700 reais (abaixo de um salário mínimo). 

Os estados em questão foram escolhidos por estarem em posições muito distintas no ranking de desenvolvimento humano municipal (São Paulo em segundo lugar e Bahia em vigesimo segundo.).(dados do IDHM de 2010). 
link : http://www.atlasbrasil.org.br/2013/pt/ranking

A faixa sálarial de menos de um sálario mínimo per capita foi selecionada por representar uma população com baixa renda mensal, consequentemente com menos recursos para o acesso a internet (tambem foi analisado o crescimento do acesso a internet desta população entre os anos de 2013 e 2015).

Ao final do projeto será possivel analisar qual é o principal meio de acesso a internet para a população escolhida (Microcomputador ou Celular).

## Conclusão 

A partir da análise dos resultados é possivel chegar em algumas conclusões :

* Houve uma regressão no número de domícilios que possuiam microcomputador entre 2013 e 2015, ao passo que o número de domicílios com internet aumentou.

Analisando as tabelas percebe-se o aumento no uso do celular para o acesso a internet.
